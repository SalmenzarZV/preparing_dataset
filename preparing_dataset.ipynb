```python
import numpy as np
import pandas as pd

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

```python
df.dtypes
```

```python
# Convert the column ‘TotalCharges’ to a numeric format, coercing errors if necessary.

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dtypes['TotalCharges']
```

```python
# Change all column names to lowercase.
for col in df.columns:
  df.rename(columns={col: col.lower()}, inplace=True)

df.columns
```

```python
df = df.replace(' ', '_', regex=True)
df
```

```python
# Convert the column ‘churn’ to integer format (no - 0, yes - 1).
df['churn'] = df['churn'].replace('No', '0', regex=True)
df['churn'] = df['churn'].replace('Yes', '1', regex=True)
df['churn'] = pd.to_numeric(df['churn'])
df
```

```python
# Check for categorical columns how many different values they have.
# Suponemos que las columnas categóricas son:
categorical_cols = ['customerid', 'seniorcitizen', 'paymentmethod']
df[categorical_cols].nunique()
```

```python
# Delete the column ‘customerid’ from the dataset
df = df.drop(columns='customerid')
df
```
